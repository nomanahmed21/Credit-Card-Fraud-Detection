```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("credit_card_fraud_10k.csv")
```

```python
df
```

```python
df = df.drop(columns="transaction_id")
```

```python
df
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoded_merchant_category = encoder.fit_transform(df[["merchant_category"]])
```

```python
encoded_merchant_category
```

```python
encoded_merchant_category = pd.DataFrame(
    encoded_merchant_category,
    columns= encoder.get_feature_names_out(['merchant_category'])
)
```

```python
encoded_merchant_category
```

```python
df = df.drop('merchant_category', axis=1)
```

```python
final_df = pd.concat([df, encoded_merchant_category], axis=1)
```

```python
final_df['is_fraud'] = final_df.pop('is_fraud')
```

```python
final_df
```

```python
final_df.to_csv("processed_data.csv")
```
